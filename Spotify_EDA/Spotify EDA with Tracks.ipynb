```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_tracks = pd.read_csv("Downloads/Spotify datasets/tracks.csv")
df_tracks.head()
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_tracks = df_tracks.sort_values("popularity", ascending = True).head(10)
sorted_tracks
```

```python
df_tracks.describe().transpose()
```

```python
most_popular_songs = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular_songs[:10]
```

```python
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[1]
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis = 1).corr(method = "pearson")
plt.figure(figsize= (14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, cmap = "viridis", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = "loudness", x = "energy", color = "r").set(title= "Loudness & Energy Correlation")
```

```python
sns.regplot(data=sample_df, y = "acousticness", x = "popularity", color = "c").set(title= "Acousticness & Popularity Correlation", )
```

```python
df_tracks['dates']= df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_dr = df_tracks.duration 
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Duration of songs for each year")
plt.xticks(rotation=90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims= (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Duration of songs for each year")
plt.xticks(rotation=60)
```

```python
df_genre = pd.read_csv("Downloads/Spotify datasets/SpotifyFeatures.csv")
df_genre.head()
```

```python
plt.title("Duration of Songs For Each Genre")
sns.color_palette("flare", as_cmap = True)
sns.barplot(x= "duration_ms", y="genre", data=df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize =(10,5))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(x = 'genre', y = 'popularity', data = famous).set(title="The 5 Most Popular Genres")
```
